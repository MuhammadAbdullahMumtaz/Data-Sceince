## Data Type Conversion

```python
# Date is in 'object', convert it into DateTime 
df['Date'] = df['Date'].astype(str).str.strip()
df['Date'] = pd.to_datetime(df['Date'], errors='coerce')
print(df['Date'].dtype)
```

```python
# 'Revenue' and 'AdSpend' columns' dtype is 'object' but that should be float64 
def parse_money(x):
    if pd.isna(x):
        return pd.NA
        
    s = str(x).lower().strip()
    value = None 

    # Detect keywords 
    if 'million' in s: 
        value = float(s.split('million')[0].strip())
        value = value * 1000000
    elif 'lakh' in s: 
        value = float(s.split('lakh')[0].strip())
        value = value * 100000
    elif 'crore' in s: 
        value = float(s.split('crore')[0].strip())
        value = value * 10000000
    else: 
        value = pd.to_numeric(x, errors='coerce')

    return value

# apply parse_money function to each element of dataframe
df['Revenue'] = df['Revenue'].apply(parse_money)
df['AdSpend'] = df['AdSpend'].apply(parse_money)

# finalyzing dtype 
df['Revenue'] = pd.to_numeric(df['Revenue'], errors='coerce')
df['AdSpend'] = pd.to_numeric(df['AdSpend'], errors='coerce')
print(df['Revenue'].dtype, df['AdSpend'].dtype)
```

```python
# 'ControversyFlag' has dtype 'object' but it should be float64

# first changin the value Yes -> 1 and No -> 0 and then convert it. 
df['ControversyFlag'] = df['ControversyFlag'].replace({'Yes':1, 'No':0})
df['ControversyFlag'] = pd.to_numeric(df['ControversyFlag'], errors='coerce')
print(df['ControversyFlag'].dtype)
```

```python
topic_keywords = {
    'Sports': ['cricket', 'football', 'hockey', 'team', 'game', 'match', 'psl', 'player', 'beat', 'defeat', 'win', 'qalandars', 'kings', 'sultans', 'babar', 'century', 'world cup', 'squash'],
    
    'Crime': ['arrest', 'police', 'murder', 'theft', 'robbery', 'suspect', 'raid', 'killed', 'violence', 'accused'],
    
    'Health': ['hospital', 'doctor', 'polio', 'dengue', 'health', 'medicine', 'covid', 'vaccine', 'patient', 'medical', 'disease', 'clinic', 'shortage', 'treatment'],
    
    'Terrorism': ['bomb', 'blast', 'terror', 'militant', 'attack', 'soldiers', 'martyred', 'ttp', 'ambush', 'hideout', 'suicide', 'explosion', 'security forces', 'foil'],
    
    'Media': ['pemra', 'channel', 'anchor', 'journalist', 'tv', 'news', 'media', 'censorship', 'editorial', 'ary', 'geo', 'dawn', 'talkshow', 'newspaper'],
    
    'Education': ['school', 'student', 'exam', 'university', 'hec', 'teacher', 'education', 'curriculum', 'fee', 'scholarship', 'matric', 'class', 'admission'],
    
    'Judiciary': ['court', 'judge', 'justice', 'lawyer', 'case', 'suo moto', 'supreme court', 'high court', 'cj', 'chief justice', 'nab', 'corruption', 'disqualif', 'proceedings', 'order', 'stay'],
    
    'Politics': ['government', 'minister', 'parliament', 'pm', 'election', 'ppp', 'pmln', 'pti', 'khan', 'assembly', 'senate', 'opposition', 'rally', 'cabinet', 'amendment', 'walkout', 'rigging', 'reform', 'party'],
    
    'Economy': ['budget', 'dollar', 'rupee', 'tax', 'imf', 'economy', 'trade', 'deficit', 'revenue', 'stock exchange', 'export', 'remittance', 'petrol', 'price', 'inflation', 'investment', 'crore', 'lakh', 'funding']
}

def is_topic_correct(headline, topic):    
    headline_lower = str(headline).lower()

    if topic not in topic_keywords:
        return False
    
    keywords = topic_keywords[topic]

    for keyword in keywords:
        if keyword.lower() in headline_lower:
            return True
    return False

df['TopicMatches'] = df.apply(lambda row: is_topic_correct(row['Headline'], row['Topic']), axis=1)

print(f"Correct: {df['TopicMatches'].sum()}")
print(f"Wrong: {(~df['TopicMatches']).sum()}")
print(f"Accuracy: {(df['TopicMatches'].sum() / len(df)) * 100:.2f}%")
```

### Outliers Detection 

```python
outlier_cols = ['Ratings', 'Revenue', 'Airtime', 'Viewership', 'Shares', 
                'AdSpend', 'SocialMediaInteractions']
outlier_summary = {}

for col in outlier_cols: 
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    outlier_count = len(outliers)
    outlier_percentage = (outlier_count / len(df)) * 100

    outlier_summary[col] = {
        'count': outlier_count,
        'percentage': outlier_percentage,
        'lower_bound': lower_bound,
        'upper_bound': upper_bound
    }
    
print(f"{'Column':<30} {'Outliers':<15} {'Percentage':<15} {'Lower Bound':<15} {'Upper Bound':<15}")
for col, info in outlier_summary.items():
    pct = f"{info['percentage']:.2f}%"
    print(f"{col:<30} {info['count']:<15} {pct:<15} {info['lower_bound']:<15.2f} {info['upper_bound']:<15.2f}")
```

The goal of this project is to find out if political bias in Pakistani media is linked to money — like how much revenue channels make or how much they earn from ads.
We want to see if channels that support the government make more money, or if all types of channels earn about the same.

**Hypotheses:**

- Channels that support the government make more money on average than neutral or opposition channels.

- Channels with stronger political bias receive more advertising money.

- Journalists who work for pro-government channels appear on TV more often.

- Serious topics like corruption, politics, and terrorism are shown less than light topics like sports or entertainment.

## Interpretation and Final Report

After cleaning and studying the data, I found some important things:

**1. Data Quality**

The dataset had lots of problems — spelling mistakes, different currency types, and many wrong topic labels.
About 88% of topics were wrongly labeled, which shows the data was either entered poorly or changed on purpose.

**2. Money and Media**

Channels that support the government actually earned a bit less in revenue and ad spending than opposition channels.
So, there's no proof in this data that pro-government media makes more money.

**3. Bias and Money**

There was almost no connection between political bias and money (revenue or ad spending).
This means channels didn’t earn more or less money just because of their political views.

**4. What Topics Were Shown**

The most talked-about topics were crime, politics, and sports.
Important issues like health, courts, and terrorism were shown less often.
This supports the idea that sensitive topics get less attention in media.

**5. Channels and Regions**

Most of the data came from Punjab and Sindh.
Express News and Samaa TV had slightly better ratings and viewership, but there wasn’t much difference in bias.

**6. Political Bias**

The average bias score was about 2.0 for all channels, which means all of them had similar bias — or the scores were adjusted (normalized) during data processing.

### Final Summary

- Pro-government media did not get more money.

- All channels had almost the same bias levels.

- Serious topics were shown less often.

- The dataset had lots of errors and missing info.

### Limitations

- The data was not real (synthetic).

- There were many mistakes and missing values.

- Topic tags were added using a simple keyword method, which is not always accurate.

- There might still be some hidden bias in the data.

### Recommendations

- Always use the same currency format (PKR) in financial data.

- Use better tools to detect bias and classify topics correctly.

- Try to work with real, clean data.

- Make clear data entry rules to avoid mistakes.
