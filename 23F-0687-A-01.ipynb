<h4 style="color:blue; font-family:'Courier New'; font-weight:bold;">Part 7</h4>

```python
dataset["Gender"] = dataset["Gender"].replace({"M": "Male", "F": "Female"})


dataset.loc[dataset["Age"] < 18, "Age"] = np.nan
dataset.loc[dataset["Age"] > 100, "Age"] = np.nan


dataset.loc[dataset["Salary"] < 20000, "Salary"] = np.nan


for col in ["Department", "Work_Location"]:
    dataset[col] = dataset[col].astype(str).str.strip().str.title()

print("Cleaned Dataset Sample:\n", dataset.head(10))
```
